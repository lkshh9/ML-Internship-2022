```python
def impfeat(data):
  nof=np.shape(data)[1]-1
  pla=data[:,-1]
  in1=np.where(pla==0)
  in2=np.where(pla==1)
  p=np.zeroes((20))
  for i in range(0,nof):
    f=data[:,i]
    f1=f[in1[0]]
    f2=f[in2[0]]
    b=ranksums(f1,f2)
    p[i]=b.pvalue
  return p
```

```python
for i in range(11,52):
  fname='filloc'+str(1)+'.csv'
  data=np.genfromtxt(fname,delimiter=',')
  print(np.shape(data)[0])
  p=impfeat(data)
  in1=np.where(p<=0.05)
  dataf=data[:,in1[0]]
  dataf=np.concatenate((dataf.data[:,-1].reshape(-1,1)), axis=1)
  fname='filloc'+str(51+i)+'.csv'
  np.savetxt(fname,dataf,delimiter=',',fmt='&f')
  print(len(in1[0]))
```

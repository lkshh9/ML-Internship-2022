LINEAR REGRESSION


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/homeprices.csv')
df.head(10)
plt.scatter(df.area,df.price)
plt.xlabel('area')
plt.ylabel('price')
```

```python
sns.set_style('whitegrid')
sns.lmplot(x ='area', y ='price', data = df)
```

```python
model = LinearRegression()
model.fit(df[['area']].values,df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5600]])
```

```python
daf = pd.read_csv('/content/homeprices_mult.csv')
daf.head(10)
```

```python
median=daf.bedrooms.median()
median
```

```python
daf['bedrooms'] = daf['bedrooms'].fillna(median)
daf
```

```python
model1 = LinearRegression()
model1.fit(daf[['area','bedrooms','age']].values,df.price)
```

```python
model1.predict([[5600,7.0,19]]) 
```

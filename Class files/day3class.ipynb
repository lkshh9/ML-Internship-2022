```python
def factorial(n):
  if n==1:
    return 1
  else:
    return n* factorial(n-1)
```

```python
factorial(5)
```

```python

def factorial(m):
  fact=1
  for i in range(1,m+1):
    fact=fact*i
  print(fact)
```

```python
factorial(5)
```
